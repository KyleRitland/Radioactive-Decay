```python
def has_decimal(number):
    
    nmbr_str_lst = list(str(number))
    
    l = len(nmbr_str_lst)
    
    binary_lst_chk = [ 0 for i in range(l) ]
    
    for i in range(l):
    
        if nmbr_str_lst[i] == '.':
            
            binary_lst_chk[i] = 1
            
        else:
            
            binary_lst_chk[i] = 0
            
    return sum(binary_lst_chk)
```

```python
def rounder_size(number):
    
    if has_decimal(number) == 0:
        
        rounder = 0
        
    elif has_decimal(number) != 0:
    
        nmbr_str_lst = list(str(number))
    
        index = nmbr_str_lst.index('.')
    
        rounder = len(nmbr_str_lst[index+1:])
    
    return rounder
```

```python
def dec_to_int_gen(number,rounder):
    
    if rounder == 0:
    
        rtrn_val = number
        
        rtrn_val_dec_l = 0
        
    elif rounder != 0:
    
        int_prt_l = len(str(number)) - ( rounder + 1 )  
        
        int_prt_val = int(''.join( list(str(number))[0:int_prt_l] ))
        
        frac_prt_lst = list(str(number))[(int_prt_l+1):]
        
        if int_prt_val == 0:
            
            rtrn_val = int(''.join( frac_prt_lst ))
            
        elif int_prt_val != 0:
            
            rtrn_val = int(''.join(( list(str(int_prt_val)) + frac_prt_lst )))
            
        rtrn_val_dec_l = len(frac_prt_lst)
            
    return rtrn_val , rtrn_val_dec_l
```

```python
def frange(initial,step,final):
    
    
    initial_rounder = rounder_size(initial)
    
    final_rounder = rounder_size(final)
    
    step_rounder = rounder_size(step)
    
    
    
    
    initial_list = dec_to_int_gen(initial,initial_rounder)
    
    initial_val = initial_list[0] 
    
    initial_dec_length = initial_list[1]
    
    step_list = dec_to_int_gen(step,step_rounder)
    
    step_val = step_list[0]
    
    step_dec_length = step_list[1]
    
    final_list = dec_to_int_gen(final,final_rounder)
    
    final_val = final_list[0]
    
    final_dec_length = final_list[1]
    
    
    
    
    
    
    
            
        
    lst_1 = [initial_dec_length , step_dec_length , final_dec_length ]
    
    lst_2 = [ initial_val , step_val , final_val ]
    
    max_dec = max(lst_1)
    
    for i in range(3):
        
        if lst_1[i] < max_dec:
            
            lst_2[i] = int(''.join(list(str(lst_2[i])) + [ '0' for i in range(max_dec - lst_1[i]) ]))
    
    
    n_m_l = len( range( min(lst_2[0],lst_2[2]) , max(lst_2[0],lst_2[2]) , lst_2[1] ) )
   
        
        
        
    
            
    number_list = [ 0 for i in range( n_m_l ) ]
        
    index = initial
    
    true_rounder = max( initial_rounder , step_rounder , final_rounder )
        
    for i in range( n_m_l ):
            
        number_list[i] = index
            
        if initial < final:
            
            index = round(index + step,true_rounder) 
                
        if initial > final:
                
            index = round(index - step,true_rounder) 
            
          
    
    return( number_list ) 
```
